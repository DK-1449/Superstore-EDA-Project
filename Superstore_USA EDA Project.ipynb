```python
#showing which order priority has most value count
plt.figure(figsize = (6,5))
ax = sns.countplot(x = 'Order Priority', data = df,color = 'brown')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12)
plt.title('count of order priority')
plt.ylabel('count of order')
plt.show()
```

# Product Category wise profit

# which product category has most shiping mode use

```python
ship_mode_counts = df.groupby(['Ship Mode','Product Category']).size().reset_index(name='Count').sort_values(by='Ship Mode', ascending=False)
ship_mode_counts
most_used_ship_mode = ship_mode_counts.loc[ship_mode_counts.groupby('Product Category')['Count'].idxmax()].sort_values(by='Ship Mode', ascending=False)
most_used_ship_mode
```

```python
plt.figure(figsize = (6,6))
ax = sns.countplot(x = 'Ship Mode' , data = df, hue = 'Product Category')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12)
plt.title('shiping mode wise product category count')
plt.ylabel('count of category')
plt.show()
```

```python
# Insight on Shipping Mode and Product Category:
# The Regular Air shipping mode is most commonly used for Office Supplies (4426 counts), followed by Technology (1772 counts), 
# indicating that Office Supplies have a higher demand for air shipping. On the other hand, Furniture relies more on Delivery Truck 
# for shipping (956 counts), possibly due to larger sizes or weight requiring ground transportation.
```

# Year wise sales trend

# What are the total sales and profit by product name?

```python
product_sales_profit = df.groupby('Product Name')[['Sales', 'Profit']].sum().reset_index()
product_sales_profit = product_sales_profit.sort_values(by='Sales', ascending=False)
product_sales_profit
```

```python
# Top 10 products by sales
top_products = product_sales_profit.head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Sales', y='Product Name', data=top_products, hue= 'Product Name',palette='magma')
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product Name')
plt.show()
```

```python
# Top 10 Customers by Sales or Profit
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='Sales', y='Customer Name', data=top_customers, hue= 'Customer Name',palette='coolwarm')
plt.title('Top 10 Customer by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Customer Name')
plt.show()
```
